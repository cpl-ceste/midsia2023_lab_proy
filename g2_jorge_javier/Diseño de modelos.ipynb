```python
#Cogemos el modelo de mejores resultados
from sklearn.linear_model import PoissonRegressor
from sklearn.ensemble import ExtraTreesRegressor, GradientBoostingRegressor, RandomForestRegressor
import xgboost
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import make_scorer, mean_squared_error
from math import sqrt
import time
from sklearn.model_selection import GridSearchCV
```

```python
def rmse_scorer(y_true, y_pred):
    mse = mean_squared_error(y_true, y_pred)
    return sqrt(mse)
```

```python
X=train[features]
y=train['SalePrice']
```

```python
from xgboost.sklearn import XGBRegressor
#generate a GradientBoostingRegressor() estimator with default parameters and random_state=0 and train it
for model in [PoissonRegressor, GradientBoostingRegressor, ExtraTreesRegressor, RandomForestRegressor, XGBRegressor]:
  if model == PoissonRegressor:
    modelo = model(alpha=50000.0, fit_intercept=True, solver='lbfgs', max_iter=100000, tol=0.0001, warm_start=False, verbose=0)
  else:
    modelo = model(random_state=0)
  #calculate cross val scoring on the train set with scoring='r2',cv=10
  #https://scikit-learn.org/stable/modules/model_evaluation.html#scoring-parameter
  scores= cross_val_score(modelo, X, y, cv=10, scoring = make_scorer(rmse_scorer))
  #print('scores -> ', scores)
  print(modelo)
  print('mean model quality ->', scores.mean())
```

### Búsqueda fina de hiperparametros del modelo

```python
import time
#https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.GradientBoostingRegressor.html#sklearn.ensemble.GradientBoostingRegressor
param_grid={'n_estimators':[50,100,150,200],
            'learning_rate': [0.1,0.075,0.05],
            'alpha': [0.5,0.9],
            'max_depth':[2,3,4],
            'min_samples_leaf':[1,3,5],
            'max_features':[1.0,2.0,3.0] }

#define the grid search method and measure the time it takes to measure and evaluate all models
time_start = time.time()
regressor = GridSearchCV(estimator=GradientBoostingRegressor() ,
                         cv=5, param_grid=param_grid,
                         scoring= make_scorer(rmse_scorer))

#train the group of models
regressor.fit(X, y)

time_end = time.time()
print("Duration: ",time_end-time_start,' secs')

#measure the scoring of the best model over the training set and get its params
regbscore=regressor.best_score_
regbparams=regressor.best_params_
print("RMSE::{}".format(regressor.best_score_))
print("Best Hyperparameters::\n{}".format(regressor.best_params_))

#calculate cross val scoring on the train set with scoring='r2',cv=10
model=regressor.best_estimator_

rmse_scores = cross_val_score(model,X, y,scoring=make_scorer(rmse_scorer),cv=5)
print('scores',rmse_scores)
print('mean score->',rmse_scores.mean())
```

# model blending

```python
#remove nulls in test data set
predict_df=test[features] #.fillna(0.0)  #in case there are some null values
y_models_list=[]  #lista de array de predicciones

modelos_name=['GradientBoostingRegressor','ExtraTreesRegressor', 'RandomForestRegressor', 'XGBRegressor','PoissonRegressor']
modelos=[GradientBoostingRegressor, ExtraTreesRegressor, RandomForestRegressor, XGBRegressor, PoissonRegressor]
for model in zip(modelos_name,modelos):
    print('Modelo:',model[0])
    if model[1] == PoissonRegressor:
      modelo = model[1](alpha=50000.0, fit_intercept=True, solver='lbfgs', max_iter=100000, tol=0.0001, warm_start=False, verbose=0)
    else:
      modelo = model[1](random_state=0)

    #important train with all data in kaggle train data set
    modelo.fit(X, y)
    y_models_list.append(modelo.predict(predict_df))

#BLENDING
#more weight to better models
weights=[0.40,0.20,0.2,0.1,0.1]  # sin poison regressor 0.35,0.25,0.2,0.2
print('total weight',sum(weights))
test['SalePrice']=weights[0]*y_models_list[0]+weights[1]*y_models_list[1]+weights[2]*y_models_list[2]+weights[3]*y_models_list[3]+weights[4]*y_models_list[4]
test['SalePrice'].head()
```
