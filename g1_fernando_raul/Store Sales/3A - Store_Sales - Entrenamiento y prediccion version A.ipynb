## Entrenamiento completo, predicción y guardado

```python
familias = data_processed.family.unique()
tiendas = data_processed.store_nbr.unique()
```

```python
full_forecast = pd.DataFrame()
empty_series_log = pd.DataFrame() # Para debugar

for t in tiendas:
  print('TIENDA: ' + str(t) + '-------------------------------------')
  first_row = data_processed.loc[(data_processed['store_nbr'] == t)].iloc[0]
  city = first_row['city']
  state = first_row['state']

  holidays = pd.DataFrame()
  for event in holidays_events.type.unique():

    # Cargar las vacaciones en un dataframe independiente para pasárselo a Prophet
    dates = holidays_events.loc[ (holidays_events['locale'] == 'Local') & (holidays_events['locale_name'] == city) | (holidays_events['locale'] == 'Regional') & (holidays_events['locale_name'] == state) |  (holidays_events['locale'] == 'National') ]
    # TO DO - añadir influencia de cada día de fiesta en los días anteriores y posteriores con lower_window y upper_window
    temp_holidays = pd.DataFrame({
        'holiday': event,
        'ds': dates.index,
    })
    holidays = pd.concat((holidays, temp_holidays))


  for f in familias:
    print('FAMILIA: ' + str(f) + '-------------------------------------')

    serie = data_processed.loc[(data_processed['family'] == f) & (data_processed['store_nbr'] == t) ]
    if serie['ds'].count() > 0:
      prophet_params = {#'growth': 'logistic',
                  'growth': 'linear',
                  'yearly_seasonality': True,
                  'weekly_seasonality': True,
                  'daily_seasonality': False,
                  'holidays': holidays,
                  'seasonality_mode': 'additive',
      }
      # Con ** desempaquetamos el diccionario de parámetros
      m = Prophet(**prophet_params)
      m.add_regressor('dcoilwtico')

      # TO DO - añadir más regresores
      # onpromotion
      # cluster

      m.fit(serie)
      future = m.make_future_dataframe(periods=16)
      # No está prediciendo los días de navidad, así que los quito de los regresores
      oil_processed_filtered = oil_processed.drop(pd.date_range(start = '2013-01-01', end = '2017-08-31' ).difference(future.ds).tolist())
      oil_processed_filtered = oil_processed_filtered.reset_index()
      future['dcoilwtico'] = oil_processed_filtered['dcoilwtico']

      forecast = m.predict(future)
      forecast_selected_columns = forecast[['ds', 'yhat']].tail(16)
      forecast_selected_columns['shop_nbr'] = t
      forecast_selected_columns['family'] = f
      full_forecast = pd.concat([full_forecast, forecast_selected_columns])
      # plot_plotly(m, forecast).show()
      # fig = m.plot(forecast)
      # a = add_changepoints_to_plot(fig.gca(), m, forecast)

    else:
      new_log = {'store_nbr':t, 'family':f}
      empty_series_log = empty_series_log.append(new_log, ignore_index=True)



# Guardamos el fichero para su tratamiento posterior
# Ahora está desordenado, habrá que ordenarlo y darle el id correcto a cada fila
full_forecast.to_csv( os.path.join( rootdir, 'forecast_temp_2.csv'), index=False, encoding='utf-8' )
empty_series_log.to_csv( os.path.join( rootdir, 'empty_series_log.csv'), index=False, encoding='utf-8' )
```
